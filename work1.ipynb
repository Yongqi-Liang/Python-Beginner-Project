```python
import turtle as t

def b(radius, angle):
    t.circle(radius, angle)

#画出矩形衬底
t.penup()
t.goto(-200,200)
t.fillcolor("#FFAB00")
t.pendown()
t.begin_fill()
t.pensize(5)
t.speed(10)
t.goto(200,200)
t.goto(200,-200)
t.goto(-200,-200)
t.goto(-200,200)
t.end_fill()
t.penup()

#画出第一笔
t.fillcolor("black")
t.goto(-60,100)
t.right(100)
t.pendown()
t.begin_fill()
t.goto(-90,-50)
t.goto(-50,-50)
t.goto(-20,100)
t.goto(-60,100)
t.end_fill()
t.penup()

t.fillcolor("black")
t.pendown()
t.begin_fill()
t.goto(25,100)
t.setheading(180)
radius=35
angle=-180
b(radius,angle)
t.right(180)
t.forward(90)
t.right(90)
t.forward(20)
t.right(90)
t.forward(80)
radius=15
angle=180
b(radius,angle)
t.forward(70)
t.end_fill()
t.left(90)
t.forward(60)
t.left(90)
#第三笔
t.fillcolor("black")
t.begin_fill()
t.forward(40)
t.right(180)
radius=22#弧度a1
angle=-85#弧度a2
b(radius,angle)#弧度a3

t.left(174)#转角

t.forward(35)
radius=50#转角b1
angle=15#转角b2
b(radius,angle)#转角b3
t.forward(2)
t.left(76)
t.forward(40)
#第三笔返航
t.left(104)
t.forward(2)
radius=50#转角b1r
angle=15#转角b2r
b(radius,angle)#转角b3r
t.right(30)
t.forward(35)

radius=30#弧度a1
angle=85#弧度a2
b(radius,angle)#弧度a3
t.forward(55)
t.left(90)
t.forward(20)
t.end_fill()
t.penup()

t.goto(10,-50)

t.setheading(0)
t.pendown()
t.fillcolor("white")
t.begin_fill()
t.width(7)
t.forward(50)
t.left(60)
t.forward(55)
t.right(140)
t.forward(47)
t.setheading(0)
t.forward(50)
t.left(210)
t.forward(55)
t.left(70)
t.forward(55)
t.right(150)
t.forward(55)
t.left(85)
t.forward(75)
t.right(155)
t.forward(60)
t.left(80)
t.forward(40)
t.end_fill()
```
